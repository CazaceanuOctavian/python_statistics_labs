### 1_Sa se calculeze coeficientii de localizare la nivel de judet ca medie a coeficientilor subactivitatilor
- Se creaza un dataframe nou in care copiem pe prima coloana judetul si pe a doua coloana media pe linie a subactivitatilor
    - luam valorile cu subactivitatile folosind iloc --> toate elementele de la coloana 2 pana la sfarsit
    - calculam media pe linie (axis = 1) pentru fiecare entry in parte
- Rezultatul se salveaza intr-un Cerinta1.csv

```python
df_coef_loc = pd.DataFrame()

df_coef_loc['Judet'] = df_energ_log['Judet']
df_coef_loc['Coef_loc'] = df_energ_log.iloc[:, 1:].mean(axis=1)

df_coef_loc.to_csv('output/Cerinta1.csv')
```

### 3_Sa se salveze judetul si subactivitatea cea mai mare din acel judet 
- Se creaza un dataframe nou in care se pun judetele 
- Se foloseste functia idxmax pe linie pentru a da fetch la numele coloanei pe care se afla valoare maxima
- Rezultatul se salveaza in Cerinta3.csv

```python
df_subactiv_max = pd.DataFrame()

df_subactiv_max['Judet'] = df_energ_log['Judet']
df_subactiv_max['Subactiv_max'] = df_energ_log.iloc[:, 1:].idxmax(axis=1)

df_subactiv_max.to_csv('output/Cerinta3.csv')
```

### 4_Sa se calculeze coeficientii medii de localizare pe nivel de regiuni 
- Facem un merge intre df_energ_loc si df_ronuts pentru a lua regiunile
- Dupa ce am facut merge facem un groupyby pe regiune unde calculam mediile astfel:
    - Facem groupby-ul folosind mean() pentru a face media valorilor numerice
    - Pentru ca groupBy-ul cu mean sa functioneze trebuie sa selectam doar coloanele numerice pentru ca nu se poate face groupby pe string-uri 
- Salvam dataframe-ul in fisierul Cerinta4.csv

```python
df_coef_reg = df_energ_log
df_coef_reg = df_energ_log.merge(df_ronuts[['IndicativJudet', 'Regiune']], left_on='Judet', right_on='IndicativJudet')
df_coef_reg = df_coef_reg.groupby('Regiune')[df_coef_reg.select_dtypes('number').columns].mean()
df_coef_reg.to_csv('output/Cerinta4.csv')
```

### 5_Sa se salveze salveze per regiune judetele cu cei mai mari coeficienti in subactivitati
- Fac un join intre df_energ_loc si df_ronuts pentru a introduce regiunile in setul de date
    - Index-ul joinului va fi reprezentat de field-ul 'Judet'
- Fac un groupby custom pe regiuni in care fiecare element rezultat va fi index-ul (judetul) unde valoarea subactivitatii este maxima  

#### Functia care se apeleaza in groupby-ul custom
- Dataframe-ul df are ca si index pe fiecare linie judetul 
- Functia se apeleaza de atatea ori cate regiuni avem (de atatea ori cate valori avem in field-ul din groupby)
    - La fiecare apel, df va fi un slice din setul de date care reprezinta o grupare pe regiuni 
- Fiecare valoare pe care o returnam reprezinta valoarea maxima dintre toate judetele per subactivitate
    - ex:
        -        c3511  c3512  c3513  c3514  c3521  c3522  c3523  c3530
        -  Judet                                                        
        -  if     0.445  0.000  0.000  0.000  0.739  0.487  1.446  0.000
        -  b      1.277  2.434  1.081  2.601  2.330  1.427  1.901  2.664
    - vom aveam b,b,b,b,b,b,b,b pentru ca b > if peste tot

```python
def custom_apply_function(df:pd.DataFrame):
    print(df)
    max = df.idxmax()
    return max
```

!! Setez ca index 'Judet' ca sa pot in custom_apply_function() sa extrag judetul !!

```python
df_aggreg = df_energ_log.merge(df_ronuts[['IndicativJudet', 'Regiune']], left_on='Judet', right_on='IndicativJudet', how='inner').set_index('Judet')
#drop useless indicativJudet column from resulted merge
df_aggreg.drop('IndicativJudet', axis=1, inplace=True)
```

```python
df_aggreg = df_aggreg.groupby('Regiune').apply(custom_apply_function, include_groups=False)
```
