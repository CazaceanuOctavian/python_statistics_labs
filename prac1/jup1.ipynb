```python
import numpy as np
from scipy.stats import kstest, shapiro, anderson, norm, chi2
import pandas as pd
```

```python
df = pd.read_csv('FreeLancerT.csv')
print(df)
```

```python
df_numerical_columns = df.select_dtypes(include='number')
print(df_numerical_columns)
```

### 1_INLOCUIREA DATELOR LIPSA CU MEDIA
- inplace = true face ca df_numerical_columns sa fie updatat cu valorile
- inplace = false returneaza un alt dataframe

```python
for column in df_numerical_columns.columns:
    column_mean = df_numerical_columns[column].mean()
    print (column_mean)
    #fill values of 0 and NaN with mean 
    df_numerical_columns[column].replace(0, column_mean, inplace=True)
    df_numerical_columns[column].fillna(column_mean, inplace=True)

print(df_numerical_columns)
```

### (varianta2) 1_INLOCUIREA DATELOR LIPSA CU MEDIA - versiunea cu masti -
- aici e acelasi aproach ca mai sus doar ca folosesc masks pentru a obtine acelasi rezultat

```python
for column in df_numerical_columns.columns:
    column_mean = df_numerical_columns[column].mean()
    print(column_mean)
    mask_is_zero = df_numerical_columns[column] == 0
    mask_is_nan = pd.isna(df_numerical_columns[column])
    #sau
    mask_is_nan = df_numerical_columns[column].isna()
    df_numerical_columns[mask_is_zero] = column_mean
    df_numerical_columns[mask_is_nan] = column_mean

print(df_numerical_columns)
```

### 2_STANDARDIZAREA DATELOR
- datele sunt standard daca au media 0 si abaterea standard 1
- pentru a standardiza un set de date pt fiecare intrare de pe coloana facem x = (x-medie)/std

```python
for column in df_numerical_columns.columns:
    df_numerical_columns[column] = (df_numerical_columns[column] - df_numerical_columns[column].mean())/df_numerical_columns[column].std() 

#print(df_numerical_columns) 
for column in df_numerical_columns.columns:
    print(f"{df_numerical_columns[column].mean():.2f} ~ {df_numerical_columns[column].std():.2f}");
```

### 3_MATRICE DE COVARIANTA/CORELATIE
- Matricele de mai jos sunt facute pe tot setul de date fara a fi grupa dupa un criteriu cum ar fi tara

```python
matrice_cov = np.cov(df_numerical_columns, rowvar=False)
matrice_corelatii = np.corrcoef(df_numerical_columns, rowvar=False)
print(matrice_cov)
print('==========')
print(matrice_corelatii)
```

### 4_MATRICE DE COVARIANTA/CORELATIE RAPORTATA LA O VALOARE
- se foloseste de exemplu daca doresc sa vad corelatia dintre variabilele numerice dupa un camp
- EX: Corelatia si covarianta intre numarul de developeri de diferite limbaje si tara din care provin
- Rezultatul este un vector de matrici. 
    - Fiecare element din vector reprezinta o matrice de MxM variabile studiate
    - Rezultatul final va fi o structura de tip QxMxM unde Q = nr tari studiate | M = nr variabile studiate 

```python
#selecteaza toate liniile de pe coloana 0
country_list = df.iloc[:, 1].unique()
#print(country_list)
number_of_unique_countires = len(df.iloc[:, 1].unique())
number_of_numerical_values = df_numerical_columns.shape[1]

corelated_cov_matrix = np.empty(shape=(number_of_unique_countires, number_of_numerical_values, number_of_numerical_values))
corelated_std_matrix = np.empty(shape=(number_of_unique_countires, number_of_numerical_values, number_of_numerical_values))

index = 0
for country in country_list:
    country_rows_mask = df['CountryCode'] == country
    df_current_county_data = df[country_rows_mask]
    df_current_numerical = df_current_county_data.select_dtypes(include='number')
    corelated_cov_matrix[index] = np.cov(df_current_numerical)
    corelated_std_matrix[index] = np.std(df_current_numerical, axis=0)
    
print(corelated_cov_matrix)
print(corelated_std_matrix)
```

### 5_TESTE DE CONCORDANTA PENTRU VERIFICAREA DISTRIBUTIEI NORMALE
- se folosesc functii din scipy 
    - kstest (Komogorov-Smirnov)
    - Shapiro 
    - Anderson
    - chi2 
- se aplica pe setul de variabile numerice (nu necesita matrici de concordanta)
- functia va stoca intr-o matrice rezultatele unde vom avea x = nr de teste facute coloane si y = nr de variabile numerice linii
    - ex: matricea va avea pe coloane cate o variabila numerica si pe linie valorile date de fiecare functie de testare
- !functia nu implementeaza testul chi2 --> se poate scrie de mana sau se poate folosi o librarie 
 

```python
statistical_tests_number = 3
number_of_numerical_values = len(df_numerical_columns.columns)
corelation_tests_matrix = np.empty(shape=(number_of_numerical_values, statistical_tests_number))

column_index = 0
# toate testele facute dau output la mai mult parametrii din care pe noi ne interesaza pvalue
# asa ca toate expresiile test_ktest[someIndex] extrag pvalue sau un indicator echivalent din outputs
for column in df_numerical_columns.columns:
    # ktest --> Komogorov-Smirnov
    test_kstest = kstest(df_numerical_columns[column], "norm")
    corelation_tests_matrix[column_index][0] = test_kstest[1] > 0.01
    # shapiro 
    test_shapiro = shapiro(df_numerical_columns[column])
    corelation_tests_matrix[column_index][1] = test_shapiro[1] > 0.01
    # anderson
    test_anderson = anderson(df_numerical_columns[column])
    # print(test_anderson)
    # testul anderson are 2 parti: test_anderson[0] --> valoare statistica si test_anderson[1] --> valori critice
    # pentru ca testul anderson sa fie valid valoare statistica trebuie sa fie mai mica decat toate valorice critice
    corelation_tests_matrix[column_index][2] = any(test_anderson[0] < test_anderson[1])
    
print(corelation_tests_matrix)
```
