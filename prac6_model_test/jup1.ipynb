### 1_Consumul mediu de alcool pe cei 5 ani la nivel de tara

```python
df_cons_mediu = pd.DataFrame()
df_cons_mediu['Country'] = df_alcohol['Country']
df_cons_mediu['Code'] = df_alcohol['Code']
df_cons_mediu['Avg'] = df_alcohol[df_alcohol.select_dtypes('number').columns].sum(axis=1)/5

df_cons_mediu.sort_values('Avg', ascending=False, inplace=True)
```

### 3_Coeficientii de variatie pe fiecare an la nivel de continent
- coef de variatie se calculeaza ca si abatere_standard/medie pentru fiecare indicator
- Se face un groupby cu regiunile
- In groupby vom aveam o functie custom care sa calculeze coeficientii per indicator

```python
def calc_coef(df:pd.DataFrame):
    print(df)
    return df.std()/df.mean()
```

```python
df_coef = df_alcohol.merge(df_coduri_tari, left_on='Country', right_on='Country', how='inner')
df_coef.drop('Country', axis=1, inplace=True)
df_coef.drop('Code', axis=1, inplace=True)
df_coef.to_csv('test')
```

```python
df_coef = df_coef.groupby('Continent').apply(calc_coef)
```
