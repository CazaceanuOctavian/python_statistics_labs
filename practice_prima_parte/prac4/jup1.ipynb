### 1_Sa se sorteze judetele descrescator in functie de ponderea persoanelor analfabete
- se creaza un dataframe nou 
- se copiaza in dataframe-ul nou judetele 
- se creaza o coloana noua 'Pondere_analfabeti' in care se calculeaza ponderea analfabetilor ca si total_pop/nr_analfabeti
- dataframe-ul se sorteaza in functie de coloana 'Pondere_analfabeti'
- se scrie intr-un fisier Cerinta1.csv

```python
df_pondere_analfabeti = pd.DataFrame()
df_pondere_analfabeti['Judet'] = df_educatie['Judet']
df_pondere_analfabeti['Pondere_analfabeti'] = df_educatie['PersoaneAnalfabete']/df_educatie.select_dtypes('number').sum(axis=1)
df_pondere_analfabeti.sort_values('Pondere_analfabeti', inplace=True, ascending=False)
df_pondere_analfabeti.to_csv('output/Cerinta1.csv')
```

### 2_Sa se sorteze judetele dupa numarul analfabetilor si persoanelor fara scoala la 100.000 de locuitori

```python
df_agreg = df_educatie.merge(df_populatie_judete, left_on='Judet', right_on='Judet', how='inner')

df_fara_scoala_si_analfabeti = pd.DataFrame()
df_fara_scoala_si_analfabeti['Judet'] = df_agreg['Judet']
df_fara_scoala_si_analfabeti['Pop_fara_studii_la_100000_oameni'] = (df_agreg['FaraScoala'] + df_agreg['PersoaneAnalfabete']) * 100000 / df_agreg['Populatie'] 
df_fara_scoala_si_analfabeti.sort_values('Pop_fara_studii_la_100000_oameni', inplace=True, ascending=False)

df_fara_scoala_si_analfabeti.to_csv('output/Cerinta2.csv')
```

### 3_Sa se extraga numarul de persoane per nivel de educatie per regiune

```python
df_agreg = df_educatie.merge(df_ronuts[['IndicativJudet', 'Regiune']], left_on='Judet', right_on='IndicativJudet')
df_agreg.drop('IndicativJudet', axis=1, inplace=True)

df_agreg = df_agreg.groupby('Regiune')[df_agreg.select_dtypes('number').columns].sum()
df_agreg.to_csv('output/Cerinta3.csv')
```

```python
df_agreg
```

### 4_Sa se extrage pentru fiecare regiune judetul cu ponderea de persoane predominanata pentru fiecare tip de ciclu de invatamant

```python
def get_significant_counties(df:pd.DataFrame):
    print(df)
    max = df.idxmax()
    return max
```

```python
df_regiuni_jud_semnificative = df_educatie.merge(df_ronuts[['Regiune', 'IndicativJudet']], left_on='Judet', right_on='IndicativJudet').set_index('Judet')
df_regiuni_jud_semnificative.drop('IndicativJudet', axis=1, inplace=True)

df_regiuni_jud_semnificative = df_regiuni_jud_semnificative.groupby('Regiune').apply(get_significant_counties, include_groups=False)

df_regiuni_jud_semnificative.to_csv('output/Cerinta4.csv')
```

### 5_Sa se calculeze numarul mediu de persoane per regiune per nivel de educatie folosind media aritmetica ponderata

```python
def calculate_weighted_mean(df:pd.DataFrame):
    print(df)
    weighted_means = np.average(df.iloc[:, :-2], weights=df['Populatie'], axis=0)
    return pd.DataFrame([weighted_means], columns=df.iloc[:, :-2].columns)
```

```python
df_ponderat = df_educatie.merge(df_populatie_judete, left_on='Judet', right_on='Judet')
df_ponderat = df_ponderat.merge(df_ronuts[['IndicativJudet', 'Regiune']], left_on='Judet', right_on='IndicativJudet').set_index('Judet')
df_ponderat.drop('IndicativJudet', axis=1, inplace=True)
```

```python
df_ponderat = df_ponderat.groupby('Regiune').apply(calculate_weighted_mean)
```
