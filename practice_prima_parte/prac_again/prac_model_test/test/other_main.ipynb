```python
df_merge = df_diversitate.merge(df_cod_local[['Siruta', 'Judet']], left_on='Siruta', right_on='Siruta')
df_merge
```

```python
df_merge['populatie'] = np.random.rand(len(df_merge)) * 10
df_merge
```

```python
def group_function(df:pd.DataFrame):
    print(df)
    print('===============')
    df_numeric = df.iloc[:, :-2]
    df_numeric = df_numeric.div(df['populatie'], axis=0)
    return df_numeric
```

```python
df_grouped = df_merge.iloc[:, 2:].groupby('Judet').apply(group_function)
df_grouped.to_csv('test')
```
