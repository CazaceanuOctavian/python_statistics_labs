```python
import numpy as np
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import linkage
```

```python
df_educatie = pd.read_csv('input/Educatie.csv')
df_ronus = pd.read_csv('input/RO_NUTS.csv')
```

```python
df_educatie.head()
```

```python
df_ronus.head()
```

```python
df_merged = df_educatie.merge(df_ronus[['IndicativJudet', 'Regiune']], left_on='Judet', right_on='IndicativJudet')
df_merged.drop('IndicativJudet', axis=1, inplace=True)
```

```python
df_merged.head()
```

```python
def grouping_function(df:pd.DataFrame):
    print(df)
    print('=======')
    std = df.std()
    mean = df.mean()
    return std/mean
```

```python
df_grouped = df_merged.iloc[:, 1:].groupby('Regiune').apply(grouping_function, include_groups = False)
```

```python
df_grouped.drop('max_indic', inplace=True, axis=1)
```

```python

df_grouped['max_val'] = df_grouped.select_dtypes(include='number').max(axis=1)
df_grouped['max_indic'] = df_grouped.idxmax(axis=1)
df_grouped
#df_grouped.drop('max_indic', axis=1, inplace=True)
```
