### 1_ Judetele sortate descrescator dupa ponderea persoanelor analfabete

```python
df_pondere_analfabeti = pd.DataFrame()
df_pondere_analfabeti['Judet'] = df_educatie['Judet']
df_pondere_analfabeti['Pondere_analfabeti'] = df_educatie['PersoaneAnalfabete']/df_educatie.iloc[:, 1:].sum(axis=1)
df_pondere_analfabeti.sort_values('Pondere_analfabeti', ascending=False, axis=0)
```

### 2_ Numarul celor analfabeti si fara scoala la 100.000 de loc

```python
df_merge = df_educatie.merge(df_pop_jud, left_on='Judet', right_on='Judet')


df_analfabet_fara_sc = pd.DataFrame()
df_analfabet_fara_sc['Judet'] = df_merge['Judet'] 
df_analfabet_fara_sc['Fara_scoala_si_analfabeti'] = (df_merge['FaraScoala'] + df_merge['PersoaneAnalfabete']) * 100000 / df_merge['Populatie'] 
df_analfabet_fara_sc.sort_values('Fara_scoala_si_analfabeti', axis=0, ascending=False)
```

### 4_ Pentru fiecare regiune, judetele cu ponderile cele mai mari pentru fiecare nivel de educatie

```python
def groupfunction(df:pd.Series):
    print(df)
    print('==============')
    return df.idxmax()
    
```

```python
df_merge_regiuni = df_educatie.merge(df_ronuts[['IndicativJudet', 'Regiune']], left_on='Judet', right_on='IndicativJudet')
df_merge_regiuni.drop('IndicativJudet', axis=1, inplace=True)
df_merge_regiuni.set_index('Judet', inplace=True)

df_merge_regiuni_grouped_max_jud = df_merge_regiuni.groupby('Regiune').apply(groupfunction, include_groups=False)

df_merge_regiuni_grouped_max_jud
```

### 5_ Sa se calculeze media persoanelor din fiecare categorie de invatamant per regiune. Se va folosi media aritmetica ponderata, ponderea fiind populatia din acea regiune

```python
def calculate_medie_ponderata(df:pd.DataFrame):
    result = np.average(df.iloc[:,:-1], weights=df['Populatie'], axis=0)
    result_formatted = pd.DataFrame([result], columns=df.iloc[:, :-1].columns)
    print(result_formatted)
    return result_formatted
```

```python
df_medie_pond_reg = df_educatie.merge(df_ronuts[['IndicativJudet', 'Regiune']], left_on='Judet', right_on='IndicativJudet')
df_medie_pond_reg.drop('IndicativJudet', axis=1, inplace=True)
df_medie_pond_reg = df_medie_pond_reg.merge(df_pop_jud[['Judet', 'Populatie']], left_on='Judet', right_on='Judet')
df_medie_pond_reg.set_index('Judet', inplace=True)
df_medie_pond_reg

weighted_avgs = df_medie_pond_reg.groupby('Regiune').apply(calculate_medie_ponderata, include_groups=False)
weighted_avgs
```
