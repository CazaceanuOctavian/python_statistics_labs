### 2_Sa se calculeze diferenta dintre presoane fara scoala la 1000 de locuitori si persoane analfabete la 1000 de locuitori per regiune

```python
df_merge = df_educatie.merge(df_ronuts[['IndicativJudet', 'Regiune']], left_on='Judet', right_on='IndicativJudet', how='inner')
df_merge.drop('IndicativJudet', axis=1, inplace=True)
```

```python
df_merge = df_merge.merge(df_populatie, left_on='Judet', right_on='Judet')
```

```python
df_merge
```

```python
df_needed = pd.DataFrame()
df_needed['Regiune'] = df_merge['Regiune']
df_needed['FaraScoala'] = df_merge['FaraScoala']
df_needed['Analfabeti'] = df_merge['PersoaneAnalfabete']
df_needed['Populatie'] = df_merge['Populatie']
df_needed.set_index('Regiune')
```

```python
df_needed.to_csv('test')
```

```python
def calc_indic(group:pd.DataFrame):
    print(group)
  # First sum up the values for each column within the group
    fara_scoala_sum = group['FaraScoala'].sum()
    analfabeti_sum = group['Analfabeti'].sum()
    populatie_sum = group['Populatie'].sum()
    
    # Calculate the difference of the sums
    indicator = (fara_scoala_sum*1000)/populatie_sum - (analfabeti_sum*1000)/populatie_sum
    
    # Return a single value for the group
    return indicator
    
```

```python
df_fin = df_needed.groupby('Regiune').apply(calc_indic, include_groups=False)
```

```python
df_fin.to_csv('test')
```

```python
def get_max_jud_fara_prelucrari(group:pd.DataFrame):
    print(group)
    for column in group.columns:
        group[column] = group[column]*1000/group['Populatie']
    print('----------TRANSFORMED------------')
    print(group)

    return group.idxmax()
```

```python
df_new_merge_f = pd.DataFrame()
df_new_merge_f['Superior'] = df_new_merge['Superior']
df_new_merge_f['Liceal'] = df_new_merge['Liceal']
df_new_merge_f['Gimnazial'] = df_new_merge['Gimnazial']
df_new_merge_f['Primar'] = df_new_merge['Primar']
df_new_merge_f['FaraScoala'] = df_new_merge['FaraScoala']
df_new_merge_f['PersoaneAnalfabete'] = df_new_merge['PersoaneAnalfabete']
df_new_merge_f['Regiune'] = df_new_merge['Regiune']
df_new_merge_f['Populatie'] = df_new_merge['Populatie']
```

```python
df_new_merge_f_final = df_new_merge_f.groupby('Regiune').apply(get_max_jud_fara_prelucrari, include_groups=False)
df_new_merge_f_final.drop('Populatie', axis=1, inplace=True)
```
